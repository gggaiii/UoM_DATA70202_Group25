```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_nor = pd.read_csv(r"C:\Github\learning_blog\Data_Science\ADS\models\model\outputs\forecast_df_nor_extended_D.csv")
df_sur = pd.read_csv(r"C:\Github\learning_blog\Data_Science\ADS\models\model\outputs\forecast_df_sur_extended_D.csv")
df_cen = pd.read_csv(r"C:\Github\learning_blog\Data_Science\ADS\models\model\outputs\forecast_df_cen_extended_D.csv")
```

```python
df_nor
```

```python
df = df_nor.head(30).copy()
df["Date"] = pd.to_datetime(df["Date"])

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(df["Date"], df["Forecast"], label='Forecast', marker='o')
plt.fill_between(df["Date"], df["Lower_CI"], df["Upper_CI"], color='gray', alpha=0.3, label='95% CI')
plt.title("Forecast with 95% Confidence Interval")
plt.xlabel("Date")
plt.ylabel("Forecast Value")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```python
# Sample structure to simulate the plotting of three datasets (df_nor, df_sur, df_cen)
# Assuming df_nor, df_sur, df_cen already exist and have "Date", "Forecast", "Lower_CI", "Upper_CI" columns

# Take top 30 rows and convert dates
df_nor_plot = df_nor.head(30).copy()
df_sur_plot = df_sur.head(30).copy()
df_cen_plot = df_cen.head(30).copy()

for df in [df_nor_plot, df_sur_plot, df_cen_plot]:
    df["Date"] = pd.to_datetime(df["Date"])

# Plotting all three datasets
plt.figure(figsize=(12, 6))

# Norte
plt.plot(df_nor_plot["Date"], df_nor_plot["Forecast"], label='Norte Forecast', marker='o')
plt.fill_between(df_nor_plot["Date"], df_nor_plot["Lower_CI"], df_nor_plot["Upper_CI"], alpha=0.2)

# Sur
plt.plot(df_sur_plot["Date"], df_sur_plot["Forecast"], label='Sur Forecast', marker='s')
plt.fill_between(df_sur_plot["Date"], df_sur_plot["Lower_CI"], df_sur_plot["Upper_CI"], alpha=0.2)

# Centro
plt.plot(df_cen_plot["Date"], df_cen_plot["Forecast"], label='Centro Forecast', marker='^')
plt.fill_between(df_cen_plot["Date"], df_cen_plot["Lower_CI"], df_cen_plot["Upper_CI"], alpha=0.2)

plt.title("Midagri Unit Price Forecasts by Region with 95% Confidence Intervals, using SARIMA model")
plt.xlabel("Date")
plt.ylabel("Forecast Value")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting only forecast lines without confidence intervals
plt.figure(figsize=(12, 6))

# Plot each region's forecast line with slightly different styles
plt.plot(df_nor_plot["Date"], df_nor_plot["Forecast"], label='Norte Forecast', marker='o', linestyle='-', linewidth=2)
plt.plot(df_sur_plot["Date"], df_sur_plot["Forecast"], label='Sur Forecast', marker='s', linestyle='--', linewidth=2)
plt.plot(df_cen_plot["Date"], df_cen_plot["Forecast"], label='Centro Forecast', marker='^', linestyle='-.', linewidth=2)

plt.title("Midagri Unit Price Forecasts by Region (No Confidence Intervals)")
plt.xlabel("Date")
plt.ylabel("Forecast Value")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```
